```python
def outlier_checking(data,colname):
    Q1=data[colname].quantile(0.25)
    Q3=data[colname].quantile(0.75)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR
    outlier=data.loc[(data[colname]<lower_limit)|(data[colname]>upper_limit)]
    return outlier

outliers = pd.DataFrame() 

for col in df.select_dtypes(include=['number']).columns: 
    outliers_in_col = outlier_checking(df, col) 
    outliers = pd.concat([outliers, outliers_in_col])

outliers.reset_index(drop=True, inplace=True)

total_rows = len(df)
outlier_count = len(outliers)
outlier_ratio = (outlier_count / total_rows) * 100

if outliers.empty:
    print("No outliers")
else:
    print(f"Number of Outliers: {outlier_count}")
    print(f"Outlier Ratio: {outlier_ratio:.2f}%")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
X=numerical_df.drop('Co2_Emissions_MetricTons', axis=1)
y=numerical_df['Co2_Emissions_MetricTons']
```

```python
X
```

```python
y
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=40)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
linear=LinearRegression()
linear.fit(X_train,y_train)
```

```python
predict = linear.predict(X_train)
predict
```

```python
print('MAE:',metrics.mean_absolute_error(y_train, predict))
print('MSE:',metrics.mean_squared_error(y_train, predict))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train, predict)))
print('R^2:',metrics.r2_score(y_train, predict))
```

```python
plt.scatter(y_train, predict)
plt.xlabel("Co2_Emissions_MetricTons")
plt.ylabel("Predicted")
plt.title("Co2_Emissions_MetricTons vs Predicted")
plt.show()
```

```python
testPredict=linear.predict(X_test)
testPredict
```

```python
print('MAE:',metrics.mean_absolute_error(y_test, testPredict))
print('MSE:',metrics.mean_squared_error(y_test, testPredict))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, testPredict)))
print('R^2:',metrics.r2_score(y_test, testPredict))
```
